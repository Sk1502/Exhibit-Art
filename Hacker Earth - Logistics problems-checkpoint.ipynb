### Outliers 

```python
#https://www.pluralsight.com/guides/cleaning-up-data-from-outliers
```

```python
Train_data.drop(Train_data[(Train_data['Weight'] < np.percentile(Train_data['Weight'],10)) | (Train_data['Weight'] > np.percentile(Train_data['Weight'],99.965))].index,inplace=True)
print(Train_data.shape)


Train_data.drop(Train_data[(Train_data['Price Of Sculpture'] < np.percentile(Train_data['Price Of Sculpture'],10)) | (Train_data['Price Of Sculpture'] > np.percentile(Train_data['Price Of Sculpture'],99.965))].index,inplace=True)
print(Train_data.shape)

Train_data.drop(Train_data[(Train_data['Cost'] < 0) | (Train_data['Cost'] > np.percentile(Train_data['Cost'],99.965))].index,inplace=True)
print(Train_data.shape)
```

```python
Train_data.describe()
```

```python
Train_data = Train_data.drop(['Customer Id','Artist Name'],axis=1)
print("Train data",Train_data.shape)
```

```python
Train_data.insert(6,'Location',Train_data['Customer Location'].str[-8:-6])
```

```python
Train_data.head()
```

```python
Train_data['Scheduled Date'] = pd.to_datetime(Train_data['Scheduled Date']) 
Train_data['Delivery Date'] = pd.to_datetime(Train_data['Delivery Date'])

Train_data['Number of days'] = Train_data['Delivery Date'] - Train_data['Scheduled Date']
Train_Date = Train_data['Number of days'].abs()
```

```python
#### Spltting dataset based on numerical ad categorical for impution


Train_data_Cat = Train_data.loc[:,['Material','International','Express Shipment','Installation Included','Transport','Fragile','Customer Information','Remote Location','Customer Location','Location']]
Train_data_Num = Train_data.loc[:,['Artist Reputation','Height','Width','Weight','Price Of Sculpture','Base Shipping Price','Cost']]      
Train_data_Date = Train_data.loc[:,['Scheduled Date','Delivery Date']]
```

### Numerical Data imputation 

```python
Train_data_Num['Artist Reputation'].replace(np.nan,Train_data_Num['Width'].median(),inplace=True)
Train_data_Num['Height'].replace(np.nan,Train_data_Num['Height'].median(),inplace=True)
Train_data_Num['Width'].replace(np.nan,Train_data_Num['Width'].median(),inplace=True)
Train_data_Num['Weight'].replace(np.nan,Train_data_Num['Weight'].median(),inplace=True)
Train_data_Num['Price Of Sculpture'].replace(np.nan,Train_data_Num['Price Of Sculpture'].median(),inplace=True)
Train_data_Num['Base Shipping Price'].replace(np.nan,Train_data_Num['Base Shipping Price'].median(),inplace=True)
```

### Min max scaling the target variable

```python
Min = Train_data['Cost'].min()
Max = Train_data['Cost'].max()
#print(Max,Min)

Train_data['Cost'] = Train_data['Cost'].abs()
Train_data['Cost']
```

```python
X = Train_data.loc[:,Train_data.columns != 'Cost']
y = Train_data['Cost']
print(X.shape,y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

### Categorical Features 

```python
from sklearn.preprocessing import OneHotEncoder 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import Normalizer

ohe = CountVectorizer()


ohe.fit(X_train['Material'].values)

X_train_Material_ohe = ohe.transform(X_train['Material'].values)
X_test_Material_ohe = ohe.transform(X_test['Material'].values)

ohe.fit(X_train['International'].values)

X_train_Intl_ohe = ohe.transform(X_train['International'].values)
X_test_Intl_ohe = ohe.transform(X_test['International'].values)


ohe.fit(X_train['Express Shipment'].values)

X_train_Shipment_ohe = ohe.transform(X_train['Express Shipment'].values)
X_test_Shipment_ohe = ohe.transform(X_test['Express Shipment'].values)

ohe.fit(X_train['Installation Included'].values)

X_train_Install_ohe = ohe.transform(X_train['Installation Included'].values)
X_test_Install_ohe = ohe.transform(X_test['Installation Included'].values)

ohe.fit(X_train['Transport'].values)

X_train_Transport_ohe = ohe.transform(X_train['Transport'].values)
X_test_Transport_ohe = ohe.transform(X_test['Transport'].values)


ohe.fit(X_train['Fragile'].values)

X_train_Fragile_ohe = ohe.transform(X_train['Fragile'].values)
X_test_Fragile_ohe = ohe.transform(X_test['Fragile'].values)

ohe.fit(X_train['Customer Information'].values)

X_train_CustInfo_ohe = ohe.transform(X_train['Customer Information'].values)
X_test_CustInfo_ohe = ohe.transform(X_test['Customer Information'].values)

ohe.fit(X_train['Remote Location'].values)

X_train_Remote_ohe = ohe.transform(X_train['Remote Location'].values)
X_test_Remote_ohe = ohe.transform(X_test['Remote Location'].values)

ohe.fit(X_train['Customer Location'].values)

X_train_CustLoc_ohe = ohe.transform(X_train['Customer Location'].values)
X_test_CustLoc_ohe = ohe.transform(X_test['Customer Location'].values)

ohe.fit(X_train['Location'].values)

X_train_Loc_ohe = ohe.transform(X_train['Location'].values)
X_test_Loc_ohe = ohe.transform(X_test['Location'].values)
```

### Numerical Features


```python
normalizer = Normalizer()

normalizer.fit(X_train['Artist Reputation'].values.reshape(-1,1))

X_train_ArtistRep_norm = normalizer.transform(X_train['Artist Reputation'].values.reshape(1,-1)).transpose()
X_test_ArtistRep_norm = normalizer.transform(X_test['Artist Reputation'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Height'].values.reshape(-1,1))

X_train_Height_norm = normalizer.transform(X_train['Height'].values.reshape(1,-1)).transpose()
X_test_Height_norm = normalizer.transform(X_test['Height'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Width'].values.reshape(-1,1))

X_train_Width_norm = normalizer.transform(X_train['Width'].values.reshape(1,-1)).transpose()
X_test_Width_norm = normalizer.transform(X_test['Width'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Weight'].values.reshape(-1,1))

X_train_Weight_norm = normalizer.transform(X_train['Weight'].values.reshape(1,-1)).transpose()
X_test_Weight_norm = normalizer.transform(X_test['Weight'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Price Of Sculpture'].values.reshape(-1,1))

X_train_Price_Sculture_norm = normalizer.transform(X_train['Price Of Sculpture'].values.reshape(1,-1)).transpose()
X_test_Price_Sculture_norm = normalizer.transform(X_test['Price Of Sculpture'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Base Shipping Price'].values.reshape(-1,1))

X_train_BaseShipping_norm = normalizer.transform(X_train['Base Shipping Price'].values.reshape(1,-1)).transpose()
X_test_BaseShipping_norm = normalizer.transform(X_test['Base Shipping Price'].values.reshape(1,-1)).transpose()

normalizer.fit(X_train['Number of days'].values.reshape(-1,1))

X_train_NumberOfDays_norm = normalizer.transform(X_train['Number of days'].values.reshape(1,-1)).transpose()
X_test_NumberOfDays_norm = normalizer.transform(X_test['Number of days'].values.reshape(1,-1)).transpose()
```

```python
X_tr = hstack((X_train_Material_ohe,X_train_Intl_ohe,X_train_Shipment_ohe, \
               X_train_Install_ohe, X_train_Fragile_ohe,X_train_CustInfo_ohe,\
               X_train_Remote_ohe,X_train_Loc_ohe,X_train_ArtistRep_norm,\
               X_train_Height_norm,X_train_Width_norm,X_train_Weight_norm,X_train_Price_Sculture_norm,\
               X_train_BaseShipping_norm,X_train_NumberOfDays_norm)).tocsr()
X_te = hstack((X_test_Material_ohe,X_test_Intl_ohe,X_test_Shipment_ohe, \
               X_test_Install_ohe, X_test_Fragile_ohe,X_test_CustInfo_ohe,\
               X_test_Remote_ohe,X_test_Loc_ohe,X_test_ArtistRep_norm,\
               X_test_Height_norm,X_test_Width_norm,X_test_Weight_norm,X_test_Price_Sculture_norm,\
               X_test_BaseShipping_norm,X_test_NumberOfDays_norm)).tocsr()

print("Final Data matrix")
print(X_tr.shape, y_train.shape)
print(X_te.shape, y_test.shape)
```
